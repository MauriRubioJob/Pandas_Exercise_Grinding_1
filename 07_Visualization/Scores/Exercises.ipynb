### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Create the DataFrame that should look like the one below.

```python
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'], 
            'female': [0, 1, 1, 0, 1],
            'age': [42, 52, 36, 24, 73], 
            'preTestScore': [4, 24, 31, 2, 3],
            'postTestScore': [25, 94, 57, 62, 70]}

data_set = pd.DataFrame(raw_data)
data_set.head()
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
plt.figure(figsize=(4,4))

plt.title("Scatter plot pre and post Score")
# plt.subplot(2,1,1)
sns.scatterplot(x=data_set.preTestScore,y=data_set.postTestScore, size=data_set.age,hue=data_set.age)
# plt.subplot(2,1,2)
# sns.scatterplot(x=data_set.postTestScore,y=data_set.preTestScore, size=data_set.age)
plt.show()
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
plt.figure(figsize=(4,4))

plt.title("Scatter plot pre and post Score")
size_pts = data_set.postTestScore*4.5
sns.scatterplot(x=data_set.preTestScore,y=data_set.postTestScore, size=size_pts,hue=data_set.female)
plt.show()
```
